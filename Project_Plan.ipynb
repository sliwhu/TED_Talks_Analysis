## Abstract  
TED is a nonpartisan nonprofit devoted to spreading ideas, usually in the form of short, powerful talks covering almost all topics — from science to business to global issues, and it costs the world nothing to watch. It is facinating how it went from nearly failing as a business to becoming a worldwide phenomenon. There are many intriguing questions behind the TED phenomenon. With such diverse range of topics, themes and speakers from various fields, which TED Talks are popular and why? There could many factors related to popularity: topics, time, speaker and even presentation skills. However, current TED Talks popularity studies are generally as shallow as listing the top viewed talks or the top speakers, deeper studies based on real data to answer the questions of what are the specific factors that relate to TED Talks popularity are lacking. Therefore, I propose this study to answer 5 research questions that relates to TED Talks popularity and reflect on how can we use the results of this study to design TED talks to make bigger impact and to appeal more people.


## 1. Introduction
TED is a nonpartisan nonprofit devoted to spreading ideas, usually in the form of short, powerful talks. TED was born in 1984 as a format for professionals in technology, entertainment and design (hence the name) to share their innovative ideas, and today it covers almost all topics — from science to business to global issues — in more than 110 languages. Now TED has become a worldwide phenomenon, attracting millions of people to watch, present, discuss and think. 

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Brief History of TED
TED wasn't this popular from the very beginning. It is born out of the founder Richard Saul Wurman's observation of a powerful convergence among three fields: technology, entertainment and design. However, TED struggled to make money for the first 6 years. In 1990, TED started to become popular. The TED Conference became an annual event in Monterey, California, attracting a growing and influential audience from many different disciplines united by their curiosity and open-mindedness -- and by their shared discovery of an exciting secret. It was still an invitation-only event back then, but it is completely open now -- anyone is welcome and encouraged to apply and attend. In 2001, media entrepreneur Chris Anderson acquired TED and took the conference nonprofit. TED was then guided by the principles that made TED further great: the inspired format, the breadth of content, the commitment to seek out the most interesting people on Earth and let them communicate their passion. Until today, TED has becoming a worldwide phenomenon, the TED family include: 
* an audio and video podcast series, [TED Talks](https://www.ted.com/about/programs-initiatives/ted-talks), in which the best TED content is released free online.
* a sister conference, [TEDGlobal](https://www.ted.com/about/conferences#h3--tedglobal), held in locations around the world.
* the [TED Prize](https://www.ted.com/about/programs-initiatives/ted-prize), which grants its winners one wish to change the world. 

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Why TED Matters and Why This Study Matters
* TED matters to the audiences

Over the past ten years TED talks have been posted online for free. There are 450,000 people log on to TED.com every day, and TED Talks are viewed some 100 million times every month - 1.2 billion times a year as of 2015. The fact that TED Talks are able to reach such huge number of people is tremendous. There is also TEDx Program which is designed to help communities, organizations and individuals to spark conversation and connection through local TED-like experiences. Launched and going online in 2009, TEDx archive has reached 100,000 talks in 2017 in 190 countries and over 2500 cities. TED organizers said they’ve received thousands of letters and emails from people telling them how one or more TED talks have impacted their lives for the better. TED has become a launching point for some interesting soul searching, creating within people's desires to learn more about things that matter outside of their own world. There is no doubt that TED Talks are impacting the world by spreading ideas, as stated in TED's mission.

* TED matters to the speakers

Not only TED speakers are influencing the audience by sharing their ideas, knowledge and experience. TED speakers' lives are also impacted dramatically for the better after their TED speeches. An interesting example from Salman Khan — founder of Khan Academy, whose goal is to bring free, world-class online education to everyone in the world — said that before his TED speech, his site had drawn only 6.9 million students. After TED, interest and demand in the courses grew like wildfire: Today, more than 140 million students take these courses online, and the site is growing by 10 million new students each month. 

With such great impacts on both the audiences and the speakers, this study will be important to understand what do the audiences care about and how do we design TED talks and invite speakers that best impact people. 

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Why TED Talks So Popular 
A few reasons are listed here. 
* TED Talks are free online.
Ted Talks have been posted online for free for the past 10 years, anyone with internet access will be able to view Ted Talks and be inspired. 
* TED Talks cover a variety of topics and given by diverse speakers
Although TED was started only for discussing technology, entertainment and design. They cover a variety of topics range from science, art, business to global issues nowadays. TED speakers are from diverse background and are working in different fields. TED's uncanny ability to connect viewers to diverse topics contribute to TED's distinctive popularity.
* TED Talks are short and focused.
TED Talks are mostly short and focused on specific topics. According to TED curator Chris Anderson who enforced an 18-minute maximum length, it was "long enough to be serious and short enough to hold people's attention". 
* TED Talks appeal to people globally
TED Talks are given and translated in hundreds of languages and the relatively new TEDx took place in thousands of cities around the world. They are not limited by geography and culture, rather, appeal to global audiences and make impact internationally.

## 2. Research Questions
### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2.1 RQ1: Define Popularity
In order to answer the research questions surrounding TED Talks popularity, popularity needs to be defined clearly first and foremost. I'd like to define popularity from the following four aspects: number of views, number of comments, number of comments per view, and number of languages. Each of these variables explains a unique aspect of popularity. Number of views suggests the general degree of attention received from the audience and maybe the most representative factor for popularity. Number of comments, on the other hand, leans more towards measuring how much the talks inspired the audience to reflect and discuss. Comments per view, combines the measurements of view and comment, defines the amount of reflection and discussion more precisely on a per view unit. Number of languages, relates more closely to the global impacts of TED talks and measures the popularity in terms of how wide the impacts on the dimension of geography and culture. 

This research question is extremely important and set constraints for the rest four questions, as all the four factors will be used to gauge and compare popularity between different TED talks.

### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 3.1 Data Sources and License

### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 3.2 Data Collection
Two datasets will be used for this project. They were downloaded from a [Kaggle Competition](https://www.kaggle.com/rounakbanik/ted-talks). These datasets contain information about all audio-video recordings of TED Talks uploaded to the official TED.com website until September 21st, 2017. The TED main dataset contains information about all talks including number of views, number of comments, descriptions, speakers and titles. The TED transcripts dataset contains the transcripts for all talks available on TED.com. The contents of each dataset are listed below:

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; TED Main Dataset

•	name: The official name of the TED Talk. Includes the title and the speaker.

•	title: The title of the talk

•	description: A blurb of what the talk is about.

•	main_speaker: The first named speaker of the talk.

•	speaker_occupation: The occupation of the main speaker.

•	num_speaker: The number of speakers in the talk.

•	duration: The duration of the talk in seconds.

•	event: The TED/TEDx event where the talk took place.

•	film_date: The Unix timestamp of the filming.

•	published_date: The Unix timestamp for the publication of the talk on TED.com

•	comments: The number of first level comments made on the talk.

•	tags: The themes associated with the talk.

•	languages: The number of languages in which the talk is available.

•	ratings: A stringified dictionary of the various ratings given to the talk (inspiring, fascinating, jaw dropping, etc.)

•	related_talks: A list of dictionaries of recommended talks to watch next.

•	url: The URL of the talk.

•	views: The number of views on the talk.

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; TED Transcripts Dataset

•	url: The URL of the talk

•	transcript: The official English transcript of the talk.
