## Abstract  
TED is a nonprofit organization born in 1984. Originally, it was set up as a platform for professionals to share their innovative ideas and the topics were restricted in technology, entertainment and design fields. Nowadays, TED has already become an unparalleled phenomenon, attracting millions of people worldwide to watch, present, discuss and think on hundreds of topics. With such diverse range of topics and speakers, which TED Talks are popular and why? Current TED Talks popularity studies are generally quite shallow as listing the top viewed talks or the top speakers, deeper studies based on real-world data to answer the question of what are the specific factors that contribute to TED Talks popularity are lacking. Therefore, I propose this study to answer 5 research questions that relates to TED Talks popularity and reflect on ways of how can we use the results of this study to understand TED talks audiences and design better TED talks in the future.

TED started with topics restricted in technology, entertainment and design, what this name stands for. However, TED topics have been much expanded over the past 30 years. Nowadays, TED topics can be ranged from science, business, art to global issues. The speakers are from diverse backgrounds as well. Therefore, people with completely different interest and desires may still find their common destination to be inspired as of TED.

According to TED curator Chris Anderson who enforced an 18-minute maximum length, it was "long enough to be serious and short enough to hold people's attention". The strategy seems very effective since it fits human nature to focus for a short amount of time. My personal observation is that short talks are more effective because people can take advantage of their fragmented. time.

## 2. Research Questions
### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2.1 RQ1: Define Popularity
In order to answer the research questions surrounding TED Talks popularity, popularity needs to be defined clearly first and foremost. I'd like to define popularity from the following four aspects: number of views, number of comments, number of comments per view, and number of languages. Each of these variables explains a unique aspect of popularity. Number of views suggests the general degree of attention received from the audience and maybe the most representative factor for popularity. Number of comments, on the other hand, leans more towards measuring how much the talks inspired the audience to reflect and discuss. Comments per view, combines the measurements of view and comment, defines the amount of reflection and discussion more precisely on a per view unit. Number of languages, relates more closely to the global impacts of TED talks and measures the popularity in terms of how wide the impacts on the dimension of geography and culture. 

This research question is extremely important and set constraints for the rest four questions, as all the four factors will be used to gauge and compare popularity between different TED talks.

TED talks are posted on line for free over the past ten years. They may be published all time over the year. In this section, I'll study the relationship of published time of TED talks vs. popularity. 

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2.3.1 Hypothesis3a TED Talks are becoming more and more popular over the past 10 years.
As time goes by, more and more people are able to access internet which should contribute to TED Talks' spread. Therefore, I hypothesize that TED Talks has becoming more and more popular over the past 10 years.

TED Talks are given all time over the years. Their popularity is mostly likely be related to the topics, speakers and presentation skills. As a result, I hypothesize that TED Talks popularity is not related to published months.

TED Talks are aiming to spread ideas, educate people and inspire impacts. Most of them are not particularly designed for the purpose of learning on specific skills or training on specific techniques. TED Talks speakers are from diverse fields with various kinds of occupations. However, certain occupations have more chances to give public speaking and therefore the speakers' presentation skills are more sophisticated. I hypothesize these speakers' talks are more popular than the rest.

TED Talks are short presentations and require sophisticated presentation skills to not only interesting enough to draw audiences' attention, but also serious enough to deliver the ideas in a short amount of time. Presentation skills are doubtless important and should be one of the reasons why some talks are more popular than others. In this section, I'll explore how presentation skills relates to popularity based on the available data. 

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2.5.1 Hypothesis5a Medium length TED Talks are more popular
The actual TED Talks lengths range from 10 minutes to 40 minutes or also. If they are too short, there may not be adequate time to convey the ideas, persuade the audience and inpspire discussions. #### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2.5.2 Hypothesis5b TED Talks that started with a question are more popular
One presentation skill is to ask questions to the audience in order to draw their attention and make them more engaged. I hereby hypothesize that TED Talks with titles containing question marks are more popular than plain titles.

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2.5.3 Hypothesis5c TED Talks presented in humorous ways are more popular
Another presentation skill is to give presentations in humorous ways that stimulate laughter. The audiences may remember the presentation contents better on interesting talks than talks that are too serious.
