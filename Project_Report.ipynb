## Abstract  
TED is a nonprofit organization born in 1984. Originally, it was set up as a platform for professionals to share their innovative ideas and the topics were restricted in technology, entertainment and design fields. Nowadays, TED has already become an unparalleled phenomenon, attracting millions of people worldwide to watch, present, discuss and think on hundreds of topics. With such diverse range of topics and speakers, which TED Talks are popular and why? Current TED Talks popularity studies are generally quite shallow as listing the top viewed talks or the top speakers, deeper studies based on real-world data to answer the question of what are the specific factors that contribute to TED Talks popularity are lacking. Therefore, I conducted this study to answer 5 research questions that relates to TED Talks popularity and reflect on ways of how can we use the results of this study to understand TED talks audiences and design better TED talks in the future. 

## 1. Introduction

#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; What do we know already about TED popularity
Why TED talks are popular? There could be many reasons contribute to its impact and influence. I have listed here a few reasons that are mostly agreed upon on the internet.

Ted Talks have been posted online for free for the past 10 years, anyone with internet access will be able to view Ted Talks and be inspired. The fact that they are free contribute to their wide reach and impact to people with even limited resources.

TED started with topics restricted in technology, entertainment and design, what this name stands for. However, TED topics have been much expanded over the past 30 years. Nowadays, TED topics can be ranged from science, business, art to global issues. The speakers are from diverse backgrounds as well. Therefore, people with completely different interest and desires may still find their common destination to be inspired as of TED.

According to TED curator Chris Anderson who enforced an 18-minute maximum length, it was "long enough to be serious and short enough to hold people's attention". ### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 2.1 RQ1: Define Popularity
In order to answer the research questions surrounding TED Talks popularity, popularity needs to be defined clearly first and foremost. I'd like to define popularity from the following four aspects: number of views, number of comments, number of comments per view, and number of languages. Each of these variables explains a unique aspect of popularity. Number of views suggests the general degree of attention received from the audience and maybe the most representative factor for popularity. Number of comments, on the other hand, leans more towards measuring how much the talks inspired the audience to reflect and discuss. Comments per view, combines the measurements of view and comment, defines the amount of reflection and discussion more precisely on a per view unit. Number of languages, relates more closely to the global impacts of TED talks and measures the popularity in terms of how wide the impacts on the dimension of geography and culture. 

This research question is extremely important and set the tone for the rest four questions, as all the four factors will be used to gauge and compare popularity between different TED talks.

TED — which stands for Technology, Entertainment and Design, naturally, my first hypothesis is that topics related to technology, entertainment and design topics are more popular.

```python
# df['tags'] are strings, apply(val) convert them to list
s = pd.Series(df['tags'].apply(eval))
# .apply(Series) converts the series of lists to a dataframe
# .stack() puts everything in one column again (creating a multi-level index)
# pd.get_dummies( ) creating the dummies
# .sum(level=0) for remerging the different rows that should be one row (by summing up the second level, only keeping the original level (level=0))
s1 = pd.get_dummies(s.apply(pd.Series).stack()).sum(level=0)
```

```python
s1.head()
```

```python
df1 = pd.concat([df['title'], df['views'], s1], axis = 1)
```

```python
df1.head()
```

##### Findings: Most given topics of all TED talks

```python
df2 = df1[df1.columns.difference(['title', 'views'])]
df2.sum(axis=0).sort_values(ascending=False)[:10]
```

##### However, the above list is most given TED talks, not necessarily indicate their popularity. Therefore, for each talk, I identify their topic tags and multiply with their number of views. Then for each topic tags, I use their total number of views of all related talks devided by the overall number of views of all talks. In this way, I'm able to compare popularity of topics in terms of view rate.

```python
sum_of_views = df1[['views']].sum()
sum_of_views
```

```python
df3 = df1[df1.columns.difference(['title', 'views'])]
#The original data column type is int8, need to change it to int64 to accomodate large numbers
df3 = df3.astype(int)
df4 = pd.concat([df1['title'], df1['views'], df3], axis = 1)
```

```python
df4.head()
```

```python
# each tag * views/total view numbers
for i in range(2, df4.shape[1]):
    for j in range(df4.shape[0]):
        if df4.iloc[j, i] != 0:
            df4.iloc[j, i]  = df4.iloc[j,i]*df4.ix[j, 'views']
```

```python
top_view_rate  = pd.DataFrame(out_view.sort_values(ascending=False)[:10])
top_view_rate.reset_index(inplace=True)
top_view_rate.columns = ['Topic', 'View Rate']
top_view_rate 
```

* Reproducibility and replicability of open research

This study will be an open research, the methodology will be freely available via the internet, along with any data or results derived from the study. The reason why I want to make it open research is to increase the scientific impact of my work, especially to help TED to appeal more audiences and make larger impact to them. As a result, reproducibility(reproducing a research study involves applying the same methods to the same data and achieving an identical result) as well as replicability(replicating a research study involves applying the same methods to new data and achieving a commensurate, confirming, or contradictory result) are important for this study. 

To fully take this human-centered data science aspect into consideration, my efforts were:
1) Clearly defined popularity concepts, four different aspects were considered, so my the audiences of my study can use results wisely based on their preference of popularity concepts.
2) All calculation metrics and constraints were clearly defined in this study, such as view rank/comment rank, view rate/comment rate. 

* Bias exist in both source data and my study design/analysis

Both bias in original data and bias in study design are exist for this study. Bias in original data is illustrated under section 1 in "Unknowns and Limitations of This Study" part. 

Bias in study design: 
1) Four definitions of popularity are proposed, which may be limited definitions. Popularity can still be assessed in other ways. 
2) Research questions are designed surrounding four aspects: topic, speaker, time and presentation skills, which may be limited. Popularity may well associated with other factors. Other study design limitations are the way I propose specific hypotheses. For example, I hypothesize 3 things in presentation skills vs popularity and they are: talk length, starting the talk with a question, presenting talk in a humourous way. There might be more presentation skills can be revealed from the transcripts! However, due to time limit of the research project, I might be bias of only selecting these three presentation skills over others, it doesn't mean these skills are more important than others.
